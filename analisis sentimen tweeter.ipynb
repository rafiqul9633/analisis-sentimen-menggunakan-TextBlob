```python
hasilanalisis = []

for tweet in hasilSearch:
    tweetProperties ={}
    tweetProperties["tanggalTweet"] = tweet.created_at
    tweetProperties["tweetId"] = tweet.id
    tweetProperties["teksTweet"] = tweet.text
    tweetProperties["pengguna"] = tweet.user.screen_name
    tweetBersih = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",tweet.text).split()) #menghapus tanda !@#$%^&&* supaya terbaca oleh textblob
    #print(tweetBersih)
    
    analysis = TextBlob(tweetBersih)
    #print(type(analysis)) #uji type teksblob
    
    #try: #diperlukan jika ada tweet yang berbahasa inggris dan indonesia
        #analysis = analysis.translate(to="en") 
    #except Exception as e:
        #print (e)
    
    #print(analysis.sentiment.polarity) # untuk mencari teks yang bersentimen +, -, atau netral
    
    if analysis.sentiment.polarity <0.0:
        tweetProperties["sentiment"] = "negatif"
    elif analysis.sentiment.polarity == 0.0:
        tweetProperties["sentiment"] = "netral"
    else:
        tweetProperties["sentiment"] = "positif"
    
    if tweet.retweet_count > 0:
        if tweetProperties not in hasilanalisis:
            hasilanalisis.append(tweetProperties)
    else:
        hasilanalisis.append(tweetProperties)
    
```

```python
sentimenPositif = [t for t in hasilanalisis if t["sentiment"]=="positif"]
sentimenNegatif = [t for t in hasilanalisis if t["sentiment"]=="negatif"]
sentimenNetral = [t for t in hasilanalisis if t["sentiment"]=="netral"]
```

```python
print("hasil sentiment")
print("positif", len(sentimenPositif),"({}%)".format(100*len(sentimenPositif)/len(hasilanalisis)))
print("netral", len(sentimenNetral),"({}%)".format(100*len(sentimenNetral)/len(hasilanalisis)))
print("negatif", len(sentimenNegatif),"({}%)".format(100*len(sentimenNegatif)/len(hasilanalisis)))
```
